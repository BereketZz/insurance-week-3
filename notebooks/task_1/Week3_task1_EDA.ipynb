```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# === LOAD DATA ===

# Load the data (change path as needed)
file_path = "MachineLearningRating_v3.txt"
df = pd.read_csv(file_path, sep="|")
```

```python
# === DATA CLEANING ===

# Convert TransactionMonth to datetime
df['TransactionMonth'] = pd.to_datetime(df['TransactionMonth'], errors='coerce')
```

```python
# Create Loss Ratio
df['LossRatio'] = df['TotalClaims'] / df['TotalPremium']
```

```python
# Clean whitespace from string columns
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
# === FILLING MISSING VALUES ===

# Fill categorical with 'Unknown'
categorical_cols = ['CrossBorder', 'Converted', 'Rebuilt', 'WrittenOff', 'Bank', 'AccountType', 'NewVehicle']
for col in categorical_cols:
    df[col] = df[col].fillna('Unknown')

# Fill numerical columns
df['CustomValueEstimate'] = df['CustomValueEstimate'].fillna(df['CustomValueEstimate'].median())

# CapitalOutstanding likely implies no outstanding amount
df['CapitalOutstanding'] = df['CapitalOutstanding'].fillna(0)

# TotalClaims and TotalPremium - very few missing, can drop
df = df.dropna(subset=['TotalClaims', 'TotalPremium'])

df = df[df['TotalPremium'] != 0]
df['LossRatio'] = df['TotalClaims'] / df['TotalPremium']

# NumberOfVehiclesInFleet - likely 1 if not part of a fleet
df['NumberOfVehiclesInFleet'] = df['NumberOfVehiclesInFleet'].fillna(1)

# Check again
print("\n✅ Missing values after cleaning:\n", df.isnull().sum().sort_values(ascending=False)[df.isnull().sum() > 0])
```

```python
# === MISSING VALUES ===
missing = df.isnull().sum().sort_values(ascending=False)
print("Missing values:\n", missing[missing > 0])
```

```python
# Fill categorical demographics with 'Unknown'
df['MaritalStatus'] = df['MaritalStatus'].fillna('Unknown')
df['Gender'] = df['Gender'].fillna('Unknown')

# Drop rows missing critical vehicle data (220 rows)
vehicle_cols = ['mmcode', 'VehicleType', 'make', 'Model', 'Cylinders', 'kilowatts',
                'bodytype', 'NumberOfDoors', 'VehicleIntroDate', 'cubiccapacity']
df = df.dropna(subset=vehicle_cols)
```

```python
# === MISSING VALUES ===
missing = df.isnull().sum().sort_values(ascending=False)
print("Missing values:\n", missing[missing > 0])
```

```python
# === UNIVARIATE ANALYSIS ===

# Plot distributions
plt.figure(figsize=(10, 5))
sns.histplot(df['TotalPremium'], bins=50, kde=True)
plt.title("Distribution of Total Premium")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['TotalClaims'], bins=50, kde=True)
plt.title("Distribution of Total Claims")
plt.show()
```

```python
# === OUTLIER DETECTION ===
plt.figure(figsize=(10, 3))
sns.boxplot(x=df['TotalPremium'])
plt.title("Boxplot of Total Premium")
plt.show()
```

```python
# === LOSS RATIO BY PROVINCE ===
loss_by_prov = df.groupby('Province')['LossRatio'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
loss_by_prov.plot(kind='bar', color='orange')
plt.title("Average Loss Ratio by Province")
plt.ylabel("Loss Ratio")
plt.xticks(rotation=45)
plt.show()
```

```python
# === TEMPORAL TRENDS ===
df_monthly = df.groupby(df['TransactionMonth'].dt.to_period('M'))[['TotalPremium', 'TotalClaims']].sum()
df_monthly.index = df_monthly.index.astype(str)
df_monthly.plot(figsize=(14, 6), marker='o')
plt.title("Total Premium & Claims Over Time")
plt.ylabel("Amount")
plt.grid(True)
plt.show()
```

```python
# === VEHICLE TYPE LOSS ===
loss_by_vehicle = df.groupby('VehicleType')['LossRatio'].mean().sort_values(ascending=False)
loss_by_vehicle.plot(kind='barh', figsize=(10, 6), color='green')
plt.title("Average Loss Ratio by Vehicle Type")
plt.xlabel("Loss Ratio")
plt.show()
```

```python
# === GENDER-BASED COMPARISON ===
loss_by_gender = df.groupby('Gender')['LossRatio'].mean()
loss_by_gender.plot(kind='bar', color='purple')
plt.title("Average Loss Ratio by Gender")
plt.ylabel("Loss Ratio")
plt.xticks(rotation=0)
plt.show()
```

```python
# === CREATIVE PLOT 2: CORRELATION HEATMAP ===
num_cols = ['TotalPremium', 'TotalClaims', 'LossRatio', 'CustomValueEstimate']
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# === CREATIVE PLOT 3: VEHICLE MAKE WITH HIGHEST CLAIMS ===
top_claims = df.groupby('make')['TotalClaims'].mean().sort_values(ascending=False).head(10)
top_claims.plot(kind='bar', figsize=(10, 5), color='red')
plt.title("Top 10 Vehicle Makes by Average Total Claims")
plt.ylabel("Avg Total Claims")
plt.xticks(rotation=45)
plt.show()
```

```python
import plotly.express as px

# Ensure TotalPremium has only positive values for plotting bubble size
zip_stats = df.groupby('PostalCode')[['TotalPremium', 'TotalClaims']].mean().reset_index()

# Remove rows where TotalPremium is zero or negative
zip_stats = zip_stats[zip_stats['TotalPremium'] > 0]

# Optional: also clean TotalClaims if needed
zip_stats = zip_stats[zip_stats['TotalClaims'] >= 0]

# Now plot safely
fig = px.scatter(zip_stats,
                 x='TotalPremium',
                 y='TotalClaims',
                 hover_name='PostalCode',
                 title="Avg Claims vs Premium per Postal Code",
                 size='TotalPremium',
                 color='TotalClaims',
                 size_max=40)

fig.show()
```
